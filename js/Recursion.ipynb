# Example of code with an infinite loop : 

- The example above shows a function with only a recusion case and continues to run itself infinitely, how does one avoid running into this issue? Base case baby.

```javascript
function multiply(arr, n) { 
    if (n <= 0) { 
        return 1;
    } else {
        return multiply(arr, n, -2) * arr[n,-1];
    }
}
```

# Base Case 

- A base case is when a fuction stops calling upon itself. This will prevent one from entering an infinite loop

* The example below presents a function which consists of a (recusion) and (base) case

```javascript
     function countdown(i) {
            console.log(i)  
            if (i <= -25) {  // base case
                return;
            } else {     // recursive case
                countdown(i - 1);
            }
        }
        
        countdown(5); 
```

- Below is another example of a function using recursion and base cases, but with increments of 1

```javascript
function countUp(int) {
    console.log(int) // Log in the input 
    if (int >= 50) { // if input is larger/equal to 50 then  // BASE CASE 
        return;  // Log and return input
    } else { // (else) if less than 50 then..
        countUp(int + 5 ); // count up from entered INT to 50  // RECURSIVE CASE
    }    
}

    countUp(0); // This is the initial call to the function
```

## The Call Stack
 - When a program calls a function, that function goes on top of the stack
   * This is similar to a stack of books. You add things one at a time. Then, when you are ready to take something off, you always take off the top item.
 
 
 ###### The example below will show the (CALL STACK) in action with the factorial function.
     - Where (factorial(5)) will be written out like --> (5! = 5 * 4 * 3 * 2 * 1) = 120

```javascript
function factorial(x) { 
    console.log(x)
    if (x == 1) { 
        return 0;
    } else { 
       return x * factorial(x-2);
        
   }
 }


factorial(5)
```
